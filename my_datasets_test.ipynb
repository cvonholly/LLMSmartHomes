```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load file
# path = "C:/Users/carlv/Documents/Projects/LLMSmartHomes/datasets/15minute_data_california/15minute_data_california/15minute_data_california.csv"
path = "C:/Users/carlv/Documents/Projects/LLMSmartHomes/datasets/15minute_data_austin/15minute_data_austin/15minute_data_austin.csv"
df_cal = pd.read_csv(path)
```

```python
# extract where 'car1' is not null
filtered_df = df_cal[df_cal["car1"].notnull()]
```

```python
filtered_df.columns
```

```python
import matplotlib.pyplot as plt

# Take 6 unique dataids and filter to one week
unique_dataids = filtered_df['dataid'].unique()[:6]
filtered_df = filtered_df[filtered_df['dataid'].isin(unique_dataids)]
filtered_df = filtered_df[filtered_df['local_15min'] < '2018-01-08 00:00:00']

plt.figure(figsize=(15, 6))

for dataid in unique_dataids:
    mask = filtered_df['dataid'] == dataid
    plt.plot(filtered_df.loc[mask, 'local_15min'],
             filtered_df.loc[mask, 'car1'],
             label=f'dataid {dataid}')

plt.legend()
plt.title('Car1 Charging Profiles for 6 Unique DataIDs')
plt.xlabel('Date')
plt.ylabel('Car1 Charging (kW)')

# Keep every 12th existing tick label
ax = plt.gca()
every_nth = 12
for i, label in enumerate(ax.xaxis.get_ticklabels()):
    if i % every_nth != 0:
        label.set_visible(False)

plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
